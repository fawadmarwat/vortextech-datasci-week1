```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    r"C:\Users\MHK\Downloads\data.csv\data.csv",
    encoding="latin1"
)
1 #DISPLAY FIRSR FIVE ROWS
df.head()
```

```python
2# DATASET INFORMATION
print("shape of dataset:", df.shape)
print()
df.info()
```

```python
3#MISSING VALUES
df.isnull().sum()
```

```python
4# DROPNA 
df.dropna(inplace=True)
df.isnull().sum()
```

*Missing Values were checking using df.isnull().sum().Rows containing missing 
values were removed using df.dropna() to make the dataset clean for analysis*.


```python
5#DUPLICATE VALUES
df.duplicated().sum()
```

```python
6#FOR REMOVE DUPLICATE VALUES
df.drop_duplicates(inplace=True)
df.duplicated().sum()
# FIX INVOICEDATA TYPE
df["InvoiceDate"]= pd.to_datetime(df["InvoiceDate"])
#FIX CUSTOMERID DATA TYPE
df["CustomerID"]= df["CustomerID"].astype(int)
#CHECK UPDATED DATA TYPE
df.info()

```

```python
top_countries = df["Country"].value_counts().head(10).reset_index()
print(top_countries.columns)
print(top_countries.head())
```

```python
# 1 BAR CHART FOR TOP 10 COUNTRIES
top_countries = df["Country"].value_counts().head(5).reset_index()
plt.figure(figsize=(12,5))
sns.barplot(data=top_countries, x="Country",y="count",color="blue")
plt.title("Top 5 Countries by Number of Orders")
plt.xlabel("Country")
plt.ylabel("Number of Orders")
plt.xticks(rotation=30)
plt.show()
```

"This BarChart shows that the top 10 countries with the highest number of orders.
 The United Kingdom has the highest number of orders.

```python
# 2 HISTOGRAM FOR QUANTITY
plt.figure(figsize=(10,5))

sns.histplot(df["Quantity"], bins=30,color="skyblue")

plt.title("Distribution of Quantity")
plt.xlabel("Quantity")
plt.ylabel("Frequency")

plt.show()
```

"Histrogram Graph shows that the distribution of product quantities in the dataset.
Most orders have low quantities.

```python
# 3 PIE CHART FOR TOP 5 COUNTRIES
top_countries = df['Country'].value_counts().head(5)
plt.figure(figsize=(8,5))
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%')
plt.title("Top 5 Countries by Number of Orders")

plt.show()
```

"This pie chart represents the percentage of orders from the top five countries."

```python
# QUICK DATA CLEAN RECAP
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(
    r"C:\Users\MHK\Downloads\data.csv\data.csv",
    encoding="latin1")
print("Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())

df.info()
```

#QUICK DATA CLEAN RECAP
"The dataset was reviewed before analysis.Missing values ,remove duplicate ,handled and the data types were verified to ensure accurate exploratory data analsis."

```python
numeric_df = df.select_dtypes(include=["number"])

plt.figure(figsize=(8,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    )


plt.title("Correlation Heatmap")

plt.show()
```

"The heatmap shows the correlation between numerical variables.Most variables have a weak 
correlation with each other.This helps identify relationships before further analysis."

```python
print(df.columns.tolist())
```

```python
# SCATTER PLOT FOR QUANTITY VS PRICE
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x="UnitPrice",y="Quantity",color="red")
plt.title("Unitprice vs Quantity")
plt.xlabel("UnitPrice")
plt.ylabel("Quantity")
plt.show()

```

"This scatter plot shows that relationship between UnitPrice and Quantity.Most products have low prices
 and are purchase in small quantity.A few outliers indicate products with unusually high values.

```python
print(df.dtypes)
```

```python
#SCATTER PLOT 2
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="CustomerID",
    y="UnitPrice",
    color="brown"
)

plt.title("Customer ID vs Unit Price")

plt.show()
```

"This graph compares CustomerID with UnitPrice.There is no stronger relationship between CustomerIDS
 and product prices.The points are randomly distributed across the chart."


```python
#BAR plot GRAPH
country_sales = (

    df.groupby("Country")["Quantity"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

plt.figure(figsize=(12,6))

sns.barplot(
    data=country_sales,
    x="Country",
    y="Quantity",
    color=""
)

plt.title("Average Quantity by Country")
plt.xlabel("Country")
plt.ylabel("Average Quantity")

plt.xticks(rotation=50)

plt.show()
```

"This bar chart compares the average quantity sold across countries.
The United Kingdom has the highest average quantity sold.
Other countries have comparatively lower sales."
